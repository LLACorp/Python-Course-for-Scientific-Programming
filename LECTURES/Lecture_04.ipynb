##  - Importing and using function packages -

A funciton **library** in a programming language is a set of scripts (in that same language or in some other ones(?)) that contain several function and class definitions, usually related to a certain topic, such that the user can call them and use them as if they had programmed them.

Examples: random, SciPy, PyGame, NumPy, matplotlib...

Usefull functions for numerical scientific methods, plotting, numerical algebra, image manipulation, web page creation, user interface generation, statistical data manipulation, videogame programming...

The main library we'll use is the Python Standard Library that comes already bundled with the core distribution, which allows the user call input/ouput functions, data structures etc (print(), input(), max(), list()...). This library is basically written in C and we just call the executables from our scripts.

This last library is already imported by default, nontheless in order to use the rest of packages we'll need to download them, integrate them in Python and then explicitly import them in the Python scripts that will use them.
```python
# let us try to correct the error as an exercice:

suposted_Identity = A_inverse @ A
for row in range(0, np.shape(suposted_Identity)[0]):
    for column in range(0, np.shape(suposted_Identity)[1]):
        if(suposted_Identity[row,column] < 1e-15):
            suposted_Identity[row,column] = 0.0

print(suposted_Identity)
```

Equation system Resolution:

Given a linear equation system, we can write it in matrix notation as A * x=b where A is the matrix with the coefficients of the variables in x, and b is the inhomogeneous coefficient part.

Then, given A and b numpy solves the system for us (if possible):

```python
A = np.array([[2, 3, 6.7], [4, 6.5, 0], [2.1, 0, 6.8]])
b = np.transpose(np.array([[4, 0, 2.9]]))
x = np.linalg.solve(A, b)
print(x)
```

```python
A @ x # shoyld give b
```

Now, imagine we want to do a linear polynomial fit with least square errors given some x data 
array and y data array. I've searched on Google:
"linear regression numpy python"
And immediately got the answer in an stackoverflow entry:
    https://stackoverflow.com/questions/6148207/linear-regression-with-matplotlib-numpy


```python
#Given the x, y data as lists or numpy arrays (it doesnt matter in this case):

x = [1.5, 2, 4.6, 8.3]
y = [3, 4.6, 8.6, 6.8]

# We obtain the polynomial fit coefficients to fit the data to y=c0+c1*x :
coefficients = np.polyfit(x, y, deg=1)
coefficients
```

Therefore, we see that the fitting line is:

    y = 0.52993806 + 3.57725396 x

If instead we wanted to make a third degree polynomial fit, we could use *np.polyfit(x,y, deg=3)*, which would have yielded an array with the four necessary coefficients to define:

    y = c0 + c1 x + c2 x^2 + c3 x^3



Now in order to evaluate new x points in the fitted linear polynomial we could deifne a function as:


```python
def linearFit( x, fittingCoeffs ):
    return fittingCoeffs[0] + fittingCoeffs[1]

linearFit(2, coefficients)
```

Alternatively, the guy answering in stackoverflow recommends creatinf a 1D polynomial using numpy with the given coefficients, such that we can evaluate whole x data vectors for instance:

```python
fittedPol = np.poly1d(coefficients) # generate the polynomial object
predictedY = fittedPol(x) # evaluate the polynomial
print(predictedY)
print(y)
```

Now imagine we are not done with just this, but we need the whole statistical information of the linear fit such as the R^2 coefficient of the fit. Then we google:

    "how to linear fit r squared python"
    
and we find that the scipy library has a statistics module where a single function returns us all the data.

https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.linregress.html

Scipy turns out to be integrated in Anaconda a swell, so we we'll just need to import it to our script (note that Scipy is very compatible with numpy, so arrays in numpy can be inputted to most functions in scipy suitable for that):


```python
from scipy import stats as scipy

x = [1.5, 2, 4.6, 8.3]
y = [3, 4.6, 8.6, 6.8]

slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(x, y)
print('Slope=',slope)
print('Intercept=',intercept)
print("R-squared=", r_value**2)
```

Now, in general, when we want to plot a particular function, we will need to create a discretization of the x axis and generate an array of evaluations of our funciton in that x axis grid. Then we plot both and voila! 

Let us as well introduce the background cartesian grid 
as well using plt.grid().

```python
# Let us define a function to plot:
def gaussian(x, sigma=1.0, mu=3.0):
    return np.exp(-0.5*((x-mu)/sigma)**2)/(sigma*np.sqrt(2*np.pi))
```

```python
# Let us build the grid:
xmin=-1
xmax=8
numberOfPoints = 500
dx= (xmax-xmin)/numberOfPoints
xPositions = [ xmin +i*dx for i in range(0, numberOfPoints)]
yPositions = [gaussian(posx) for posx in xPositions]
```

```python
plt.plot(xPositions, yPositions)
plt.xlabel("Position")
plt.ylabel("Probability")
plt.title("My Gaussian")
plt.show()
```

```python
plt.plot(xPositions, yPositions)
plt.xlabel("Position")
plt.ylabel("Probability")
plt.title("My Gaussian")
plt.grid(True)
plt.show()
```

Let us explore the errorbars as well, very useful in scientific plotting:

```python
# we make up the x, y values
x = np.arange(0, 4*np.pi, 0.3)
y = np.sin(x) + 0.6 * np.random.randn(len(x))

# we invent some errors for each value in x and y

y_error = 0.6
x_error = 0.1*np.sqrt(x)
```

The plt.errorbar() function takes first the x, y vectors that will tell the function where to plot the error bars in the graph (usually over the data points that represent them), afterwards it takes the error in the x axis and the one in the y axis. If the additional argument fmt='none' is added, then the function will only plot the error bars; else, the points will be drawn as well, which perhaps can more conveninetly be done with the plt.plot() function instead.

My recommendation: draw them separatedly so you can take profit of all the customization aspects we've seen so far.

```python
plt.errorbar(x, y, yerr=y_error, xerr=x_error, fmt='none')
plt.plot(x,y, 'or', markersize=2.5)
plt.show()
```

Errorbars can be tuned in many different ways as well, here you have a sample: the "capsize" argument lets you picture shorter or longer caps in the error bars; "ecolor" stands for error bar color; elinewidth stands for the line width of the error bars...

See the help (?) for more information or look for it in the net.

## 3D Data Visualization

Last but not least, let us have a quick taste of the 3D surface plots, contour plots and heat maps:

### Contour Maps

It is often usefull to visualize 3D data in 2D using conoturs or color-coded regions (heat maps). Three functions in matplotlib are useful for that: plt.contour for simple contour plots, plt.contourf for filled contour plots and plt.imshow for images.

Let us use as an example the function z=f(x,y) of a "two mountain" 2D Gaussian distribution:

```python
def f(x,y):
    sigmax=1.0
    sigmay1=1.0
    sigmay2=2.0
    mux=0
    muy1=4.0
    muy2=-4.0
    return -0.05 + (1/(sigmax*np.sqrt(2*np.pi)))*np.exp(-0.5*((x-mux)/sigmax)**2) * ((1/(sigmay2*np.sqrt(2*np.pi)))*np.exp(-0.5*((y-muy2)/sigmay2)**2) + (1/(sigmay1*np.sqrt(2*np.pi)))*np.exp(-0.5*((y-muy1)/sigmay1)**2)) 
```

Contour plots using *plt.contour()* require three arguments:

- A grid for the x axis values
- A grid for the y axis values
- A grid for the z axis values

The x and y values will represent the possible x,y arguments of the function, while z will be the output of the function, which will give rise to the contour levels.

The easiest way to prepare the data is to use the *np.meshgrid()* function, which given a 1D x array and a 1D y array meshes them into a 2D grid:

```python
x = np.linspace(-8, 8, 100)
y = np.linspace(-10, 8, 100)

X,Y = np.meshgrid(x,y)
```

Note that X and Y are each a 2D matrix with the values in 1D repeated all over the matrix in a way that taking all the [i,j]-th element from both of them returns all the possible pairs of (x,y) in the grid.

Therefore, evaluating f(X,Y) will return a matrix with the z values corresponding to each possible pair (x,y) in the grid:

```python
Z = f(X,Y)
plt.contour(X, Y, Z)
plt.show()
```

If nothing else passed as argument, it will just plot some default number of contour plots with a transparency degradation proportional to the height of z. Higher in z the more opaque the contour appears.

If we set the option of single colour 'colors', then the negative values will appear as dashed lines:

```python
plt.contour(X, Y, Z, colors='black')
plt.show()
```

Alternatively, the contours can be color-coded by specifying the colourmap in the 'cmap' argument.

We note that if we introudce an integer as the 4-th argument, this will impose the number of contour lines to plot:

```python
numberOfContourCurves = 20
plt.contour(X, Y, Z, numberOfContourCurves, cmap='RdGy')
plt.show()
```

Here we chose the RdGy (Red-Gray) colormap, convenient for centered data, but matplotlib has a very wide range of colormaps available, look for them in the net.

Now, probably we'd like to have a colourbar in order to have the numerical legend of the colours. This can be easily added with plt.colorbar()

```python
numberOfContourCurves = 20
plt.contour(X, Y, Z, numberOfContourCurves, cmap='RdGy')
plt.colorbar()
plt.show()
```

As we can see, the blacker regions are the peaks while the red ones are the depths.

Now, what if we'd like to avoid the blank spaces between the contour lines: we can make the same plot but with a filled contour by using *plt.contourf()* funciton, which uses the same syntax as plt.contour():

```python
numberOfContourCurves = 10
plt.contourf(X, Y, Z, numberOfContourCurves, cmap='RdYlBu')
plt.colorbar()
plt.show()
```

The discretization of the levels may be disturbing sure, which could be faced by using many many conotur levels:

```python
numberOfContourCurves = 100
plt.contourf(X, Y, Z, numberOfContourCurves, cmap='RdYlBu')
plt.colorbar()
plt.show()
```

Nonetheless, there exists a more efficient way of doing this, by using the *plt.imshow()* function which interprets a 2D matrix data as an image.

The funciton takes as arguments:
- the image matrix (Z in our case)
- the extents of the x,y grid [xmin, xmax, ymin, ymax]
- the origin of the xy grid
- The colormap

Note that the aspect-ratio, by default, will be adjusted to match the number of elements in x and the ones in y. we can use *plt.axis(aspect='image')* to match x and y units.

Also, we can use before plotting anything the command *plt.figure(figsize=(,))* to make a bigger plot:


```python
plt.figure(figsize=(10,10))

plt.imshow(Z, extent=[-8, 8, -10, 8], origin='lower',cmap='RdYlBu')
plt.colorbar()
plt.axis(aspect='image');
```

Finally, we could mix both ideas by using in the background a colour plot with a lower opacity (measured by hte 'alpha' argument) and over it sketching labelled contours:

```python
plt.figure(figsize=(10,10))

contours = plt.contour(X, Y, Z, 7, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z, extent=[-8, 8, -10, 8], origin='lower',
           cmap='RdGy', alpha=0.7)

plt.colorbar();
```

Plotting scattered points or lines can be done with the *ax.plot3D()* function, which takes data very much like the 2D plt.plot() function. Here we'll plot a spiral line representing the motion of a 2D particle in time (z axis will be time). First we'll produce only 100 points to illustrate the scatter plot, and then 1000 to see the solid line plot:

```python
ax = plt.axes(projection='3d')
t_data= np.linspace(0, 15, 100) # just 10 time points
x_data = np.sin(t_data)
y_data = np.cos(t_data)

ax.plot3D(x_data, y_data, t_data, 'or')
plt.show()
```

```python
ax = plt.axes(projection='3d')
t_data= np.linspace(0, 15, 1000) # just 10 time points
x_data = np.sin(t_data)
y_data = np.cos(t_data)

ax.plot3D(x_data, y_data, t_data, '-r')
plt.show()
```

For the scatter plot a depth feeling may be imprinted by using the *scatter3D()* function instead of the standard plot3D. This function wil give the dots a transparency according to the depth they are in with respect to the camera. Using a custom color map with the 'cmap' argument we can even give them a colour according to the height in z. The c argument will take the data according to which the colors will be depicted.

```python
ax = plt.axes(projection='3d')
t_data= np.linspace(0, 15, 100) # just 10 time points
x_data = np.sin(t_data)
y_data = np.cos(t_data)

ax.scatter3D(x_data, y_data, t_data, c=t_data, cmap='winter')
plt.show()
```

In order to plot a 3D contour plot there is the *ax.contour3D()* function that takes exactly the same arguments as the 2D contour plot. Let us plot in 3D the same function f(x,y) we tried to imagine using the 2D contours:

```python
def f(x,y):
    sigmax=1.0
    sigmay1=1.0
    sigmay2=2.0
    mux=0
    muy1=4.0
    muy2=-4.0
    return -0.05 + (1/(sigmax*np.sqrt(2*np.pi)))*np.exp(-0.5*((x-mux)/sigmax)**2) * ((1/(sigmay2*np.sqrt(2*np.pi)))*np.exp(-0.5*((y-muy2)/sigmay2)**2) + (1/(sigmay1*np.sqrt(2*np.pi)))*np.exp(-0.5*((y-muy1)/sigmay1)**2)) 

x = np.linspace(-8, 8, 100)
y = np.linspace(-10, 8, 100)

X,Y = np.meshgrid(x,y)

Z=f(X,Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.contour3D(X, Y, Z, 50, cmap='RdYlBu')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```

The initial viewing angle can be set with the function *ax.view_init()* which takes two arguments: the two spherical angles of the camera. This could as well be adjusted by using the interactive matplotlib mode.

```python
ax.view_init(60, 35)
fig
```

Finally it may be interesting to plot the figure using a wireframe or a surface plot. These can be achieved using the functions *ax.plot_wireframe()* and *plt.plot_surface()* which take very similar arguments to the contour plot. A difference is that both take a pair of arguments called "rcount" and "ccount" which stand for the number of rows and columns to "poligonalize":

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z, rcount=15, ccount=15)
ax.set_title('wireframe');
```

The surface plot is like the wireframe but with the polygons filled with a colour. Theefore, if we set many many rcounts and ccounts, we could make the plot seem a continous function.

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rcount=25, ccount=25,
                cmap='RdYlBu')
ax.set_title('surface');
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rcount=100, ccount=100,
                cmap='RdYlBu')
ax.set_title('surface');
```
